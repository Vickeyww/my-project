```python
# Hyperparameters to be checked.
parameters = {'C':[0.0001, 0.001, 0.01, 1, 0.1, 10, 100, 1000], 'penalty':['none','l2'] ,
              'solver':['lbfgs','sag','saga','newton-cg']}

# Logistic Regression estimator.
lr=LogisticRegression(random_state=42)

# GridSearchCV estimator.
m1 = GridSearchCV(lr, parameters, cv=5, n_jobs=-1, verbose=1)

# Train the GridSearchCV estimator and search for the best parameters.
m1.fit(X_train_scaler,y_train)
```

### PCA Analysis

#### Train a new PCA object

```python
# standardize features in training set

scaler = StandardScaler()

X_train_normalized = scaler.fit_transform(X_train)
```

```python
# 1. train a PCA object,retuen all PCs
from sklearn.decomposition import PCA

pca1 =  PCA()   # n_components = None (default), return all PCs

X_pca1 = pca1.fit_transform(X_train_normalized)  # fit and transform the data
```

#### Find the numbers of PCs

```python
# 2. choose number of PCs to use

# create a function so that we can use it later

def scree_plot(explained_var):
    plt.figure(figsize=(30,8))
    plt.plot(explained_var, 's--', linewidth = 4, markersize = 8);     # dashed line with square marker
    plt.xlabel('Principal Components')
    plt.ylabel('Explained Variance')
    
    ticks = np.arange(len(explained_var))                             # index of each PC: 0, 1, 3, 4
    labels = ['PC' + str(i) for i in np.arange(1, len(explained_var)+ 1)]   
    plt.xticks(ticks = ticks, labels = labels)
    plt.title('Screen Plot');
```

```python
print("No. of PCs with variance > 1: ", sum(pca1.explained_variance_ > 1))

print("% variance explained by top 5 PC: ", np.cumsum(pca1.explained_variance_ratio_)[4])

scree_plot(pca1.explained_variance_)  
```

We select top 5 Pcs to build our following 3 models

```python
 # 3. visualize the coefficents for top 9 PCs and all 24 features
def PCA_loadings(coefs, features_names):
    plt.figure(figsize = (30, 10))
    # Display an array as a matrix in a new figure window (fignum = 0 means use the current axes)
    plt.matshow(coefs, cmap='BrBG', fignum=0)   
    plt.xticks(ticks = range(len(features_names)), labels = features_names, rotation=30, ha='left')
    plt.yticks(ticks = np.arange(len(coefs)), labels= ['PC' + str(i) for i in np.arange(1, len(coefs)+ 1, 1)] )    
    plt.xlabel("Features")
    plt.ylabel("Principal components")  
    plt.colorbar(location = 'bottom');
```

```python
coefs = pca1.components_       

PCA_loadings(coefs[:5,:] , df.columns.drop('IsDefaulter'))   
```

```python
# 3. biplot
def biplot(scores, coefs, features_names, target = None):   
    
    x = scores[:,0]     # PC1
    y = scores[:,1]     # PC2
    n = len(features_names)   # no. of features
    
    scalex = 1.0/(x.max() - x.min())   # scale scores so that scaled scores' range is [-1, 1]
    scaley = 1.0/(y.max() - y.min())
    
    plt.figure(figsize = (12, 8))
    
    if target is None:
        scatter = plt.scatter(x * scalex, y * scaley)     # scatter plot for scaled pc scores
    else:
        scatter = plt.scatter(x * scalex, y * scaley,c = target)   # map color to target labels 
    
    for i in range(n):
        # add arrows, the head is defined by each feature' coef with PC1 and PC2
        plt.arrow(x = 0, y = 0, dx = coefs[0,i], dy = coefs[1,i], width=.005, color='r', alpha=0.5)  
        
        # add texts for arrows, text position is defined by each feature' coef with PC1 and PC2
        plt.text(x=coefs[0,i]*1.15, y=coefs[1,i]*1.15, s = features_names[i], color='g', ha='left', va='bottom')
    
    plt.legend(*scatter.legend_elements())   # add legends
    
    xleft, xright = plt.xlim()             # get current axis limit
    plt.xlim(xleft * 1.2, xright * 1.2)    # expand the limit a bit
    yleft, yright = plt.ylim()
    plt.ylim(yleft * 1.2, yright * 1.2)
    
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.grid()  # add some grid lines
```

```python
def get_pre_rec_f1(model,X_test,y_test):
    y_pred = model.predict(X_test)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    
    Accuracy=(tp+tn)/(tn+fp+fn+tp)
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    F1 = 2 * (precision * recall) / (precision + recall)
    
    print(f'Accuracy：{Accuracy:.3f}\nPrecision:{precision:.3f}\nRecall:{recall:.3f}\nF1 score:{F1:.3f}')
```

```python
# Calculate precision and recall of Logistic Regression model

print('Logistic Regression model on test data:')
get_pre_rec_f1(grid1.best_estimator_, X_test, y_test)
```

```python
from tensorflow import keras
from tensorflow.keras import layers
model1 = keras.Sequential([
    layers.Dense(units =50, activation="tanh"),        
    layers.Dropout(rate = 0.1),                         
    layers.Dense(units = 50, activation="tanh"),
    layers.Dropout(rate = 0.1),
    layers.Dense(units = 50, activation="tanh"),
    layers.Dropout(rate = 0.1),
    layers.Dense(units = 50, activation="relu"),
    layers.Dropout(rate = 0.01),
    layers.Dense(units = 50, activation="relu"),
    layers.Dropout(rate = 0.01),
    layers.Dense(units = 50, activation="relu"),
    layers.Dropout(rate = 0.001),
    layers.Dense(units = 50, activation="relu"),
    layers.Dropout(rate = 0.001),
    layers.Dense(units = 1, activation="sigmoid")      # output probability of 1
    ])
```

```python
model1.compile(optimizer="rmsprop",
              loss="binary_crossentropy",    
              metrics=["accuracy"])
```
